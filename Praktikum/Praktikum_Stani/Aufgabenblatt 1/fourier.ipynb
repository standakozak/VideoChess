```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def show_fourier(amplitudes, period, start=-4, end=4):
    ts = np.linspace(start, end, 100)
    approximation_steps = lambda t: [amplitude * np.sin((i * 2 * np.pi * t) / period) for i, amplitude in enumerate(amplitudes)]
    y = [sum(approximation_steps(t)) for t in ts]
    f, (ax1, ax2) = plt.subplots(1, 2)
    ax1.plot(ts, y)
    ax2.vlines(x=[i for i in range(0, len(amplitudes))], ymin=0, ymax=amplitudes)
    ax2.scatter([i for i in range(0, len(amplitudes))], amplitudes, marker="o")
```

```python
# Beispiel 1
n_sinusoids = 10
amplitudes = [0, 1, 0, 0, 0]
show_fourier(amplitudes, period=2*np.pi)
```

```python
# Beispiel 2
n_sinusoids = 10
amplitudes = [0] + [4 / (i * np.pi) if i % 2 == 1 else 0 for i in range(1, n_sinusoids)]
show_fourier(amplitudes, period=2)
```

```python
# Beispiel 3
n_sinusoids = 50
amplitudes = [0] + [2 / (i * np.pi) if i % 2 == 1 else 0 for i in range(1, n_sinusoids)]
show_fourier(amplitudes, period=4)
```
